```python
import numpy as np
import pandas as pd
from tqdm import tqdm_notebook
```

```python
sampling_seed = 12358
```

```python
df0 = pd.read_csv('data/training/center_0.csv').set_index('Unnamed: 0')
df0.head()
```

```python
def sample_df(df, random_seed):
    df_tumor = df[df['tumorLabel'] == True]
    df_normal = df[df['tumorLabel'] == False]
    tumor_count = len(df_tumor)
    # sample randomly the same amount of normals
    df_normal_sample = df_normal.sample(tumor_count, random_state=random_seed)
    # combine and shuffle
    df_combined = pd.concat([df_tumor, df_normal_sample], ignore_index=True)
    # this shuffles
    return df_combined.sample(frac=1, random_state=random_seed + 1)
```

```python
df0_sample = sample_df(df0, sampling_seed)
df0_sample.describe(include='all')
```

```python
df1 = pd.read_csv('data/training/center_1.csv').set_index('Unnamed: 0')
df1_sample = sample_df(df1, sampling_seed)
df1_sample.describe(include='all')
```

```python
df2 = pd.read_csv('data/training/center_2.csv').set_index('Unnamed: 0')
df2_sample = sample_df(df2, sampling_seed)
df2_sample.describe(include='all')
```

```python
df3 = pd.read_csv('data/training/center_3.csv').set_index('Unnamed: 0')
df3_sample = sample_df(df3, sampling_seed)
df3_sample.describe(include='all')
```

```python
df4 = pd.read_csv('data/training/center_4.csv').set_index('Unnamed: 0')
df4_sample = sample_df(df4, sampling_seed)
df4_sample.describe(include='all')
```

Save splits 

```python
# combine training sets
df_training = pd.concat([df0_sample, df2_sample, df3_sample], ignore_index=True)
df_training = df_training.sample(frac=1, random_state=sampling_seed)
df_training.describe(include='all')
```

```python
#split validation set
df_val_1 = df1_sample[len(df1_sample)//2:]
df_val_2 = df1_sample[:len(df1_sample)//2]
```
